```python
import numpy as np
import pandas as pd
import japanize_matplotlib
import matplotlib.pyplot as plt
import os

%matplotlib inline
os.chdir("../")
```

```python
def calc_stddev(data: list):
    return np.std(data, ddof=1)
```

```python
def histplot(data, title:str, xlabel:str, bins:int=20):
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)

    ax.hist(data, bins=bins, range=(0, 1))
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel("度数")
```

```python
df1 = pd.read_csv("csv/GReeeeN_tree_similarity_by_subtree_count_20231119.csv")
df2 = pd.read_csv("csv/ヨルシカ_tree_similarity_by_subtree_count_20231119.csv")
data = pd.concat([df1, df2])
rates = pd.concat([data.numerator_A / data.denominator_A, data.numerator_S / data.denominator_S], axis=0)

print(calc_stddev(rates))
histplot(rates, title="部分木の個数計算による類似度の頻度分布（重みなし）", xlabel="類似度（階級）")
```
