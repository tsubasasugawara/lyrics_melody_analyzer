```python
import numpy as np
import pandas as pd
import japanize_matplotlib
import matplotlib.pyplot as plt
import os

%matplotlib inline
os.chdir("../")
```

```python
def calc_stddev(data: list):
    return np.std(data, ddof=1)
```

```python
def histplot(data, xlabel:str, bins:int=30):
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)

    ax.hist(data, bins=bins, range=(0, 1))
    ax.set_xlabel(xlabel)
    ax.set_ylabel("度数")
```

# 単語の一致率の度数分布

```python
df1 = pd.read_csv("csv/GReeeeN_wmr_20231119.csv")
df2 = pd.read_csv("csv/ヨルシカ_wmr_20231119.csv")
data = pd.concat([df1, df2])
rates = pd.concat([data.numerator_A / data.denominator_A, data.numerator_S / data.denominator_S], axis=0)

print(calc_stddev(rates))
histplot(rates, xlabel="類似度（階級）")
```

```python
df1 = pd.read_csv("csv/GReeeeN_ts_20231119.csv")
df2 = pd.read_csv("csv/ヨルシカ_ts_20231119.csv")
data = pd.concat([df1, df2])
rates = pd.concat([data.numerator_A / data.denominator_A, data.numerator_S / data.denominator_S], axis=0)

print(calc_stddev(rates))
histplot(rates, xlabel="類似度（階級）")
```
