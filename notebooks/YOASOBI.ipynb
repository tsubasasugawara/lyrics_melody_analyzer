```python
import os
import pandas as pd
from scipy import stats
import notebook as nb
%matplotlib inline


os.chdir("../")

df = pd.read_csv("csv/YOASOBI_20240204.csv")
```

# 単語の一致率（YouTube)

```python
rates_wmr, xy_min_wmr, xy_max_wmr = nb.preprocess(df, "csv/YOASOBI_wmr_20240204.csv")
nb.scatter(rates_wmr, xy_min_wmr, xy_max_wmr, nb.YOUTUBE, "Aメロの単語の一致率", "サビの単語の一致率", "images/YOASOBIにおける単語の一致率と人気度の関係（YouTube）.svg")
```

## t検定(サビ，軸回転なし)
- 帰無仮説：サビの単語の一致率の平均は等しい
- 有意水準：0.05

```python
sorted_rates_wmr = nb.preparing_ttest(rates_wmr, nb.YOUTUBE)
stats.ttest_ind(sorted_rates_wmr["pop"][nb.RATE_S], sorted_rates_wmr["unpop"][nb.RATE_S], equal_var=False)
```
