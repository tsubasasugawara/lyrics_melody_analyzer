```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import os
os.chdir("../")
```

```python
pop = pd.read_csv("csv/word_match_rate_pop.csv")
print(pop)
```

```python
unpop = pd.read_csv("csv/word_match_rate_unpop.csv")
print(unpop)
```

```python
fig = plt.figure()

ax = fig.add_subplot(1,1,1)

ax.scatter(pop["Aメロ"], pop["サビ"], c='red')
ax.scatter(unpop["Aメロ"], unpop["サビ"], c='blue')

ax.set_title('Word match rate')
ax.set_xlabel('A')
ax.set_ylabel('S')

fig.show()
```
